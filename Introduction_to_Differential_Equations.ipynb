# Introduction to Differential Equations with Python

Differential equations are mathematical equations that relate a function with its derivatives. They are widely used in engineering, physics, biology, economics, and many other fields to model dynamic systems and processes.

In this tutorial, we'll learn the basics of differential equations and see how to work with them using Python.

---

## 1. What is a Differential Equation?

A **differential equation** is an equation that involves an unknown function and its derivatives. It describes how a quantity changes with respect to another, often time or space.

**General form:**
$$
F\left(x, y, \frac{dy}{dx}, \frac{d^2y}{dx^2}, ..., \frac{d^ny}{dx^n}\right) = 0
$$

### Example
$$
\frac{dy}{dx} + 2y = e^x
$$

In this equation:
- $ y $ is the unknown function of $ x $.
- $ \frac{dy}{dx} $ is the first derivative of $ y $ with respect to $ x $.

---

## 2. Order and Degree of Differential Equations

### **Order**

- The **order** of a differential equation is the order of the highest derivative present in the equation.

    **Example:**
    $$
    \frac{d^2y}{dx^2} + 3\frac{dy}{dx} + 2y = 0
    $$
    Here, the highest derivative is $ \frac{d^2y}{dx^2} $, so the order is **2**.

### **Degree**

- The **degree** (when defined) is the power of the highest order derivative, after the equation is made polynomial in derivatives.

    **Example:**
    $$
    \left(\frac{d^2y}{dx^2}\right)^3 + \frac{dy}{dx} + y = 0
    $$
    Here, the highest order derivative is $ \frac{d^2y}{dx^2} $ and its power is 3, so the degree is **3**.

    **Note:** Degree is defined only when the equation can be written as a polynomial in derivatives.

---

## 3. Working with Differential Equations in Python

We'll use the `scipy` library, specifically `scipy.integrate.solve_ivp` or `odeint`, to solve ordinary differential equations (ODEs).

Let's solve a simple first-order ODE:

$$
\frac{dy}{dx} = -2y, \quad y(0) = 1
$$

### Step 1: Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

### Step 2: Define the Differential Equation

The equation $ \frac{dy}{dx} = -2y $ can be represented as a Python function:

```python
def dydx(x, y):
    return -2 * y
```

### Step 3: Set Initial Conditions and Range

```python
x_span = (0, 5)          # Solve from x=0 to x=5
y0 = [1]                 # Initial condition y(0) = 1
```

### Step 4: Solve the ODE

```python
solution = solve_ivp(dydx, x_span, y0, t_eval=np.linspace(0, 5, 100))
```

### Step 5: Plot the Solution

```python
plt.plot(solution.t, solution.y[0], label='y(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Solution of dy/dx = -2y with y(0)=1')
plt.legend()
plt.grid(True)
plt.show()
```

## 4. Summary

- A differential equation involves an unknown function and its derivatives.
- **Order**: Highest derivative in the equation.
- **Degree**: Power of the highest order derivative (when polynomial).
- Python, with libraries like `scipy`, makes it easy to solve and visualize solutions to differential equations.
